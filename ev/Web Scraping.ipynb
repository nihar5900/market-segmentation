```python
import undetected_chromedriver as uc #For using Chrome browser
from selenium.webdriver.common.by import By #For locating element
import time #For adding delays
from tqdm.notebook import tqdm #For progress bar
import pandas as pd

name, price, rating, speed, charge, wight, fl_charge, cc, milage, power = [], [], [], [], [], [], [], [], [], []

## Set Chrome Options
options = uc.ChromeOptions()
options.add_argument("--incognito")
options.add_argument('--disable-popup-blocking')
options.add_argument("--force-device-scale-factor=0.8")
options.headless=False

driver = uc.Chrome(options=options, version_main=116)
#Maximize window
driver.maximize_window()
#Open target website
# normal https://www.bikewale.com/electric-bike/?sortField1=modelPopularity&sortOrder1=desc&budget=0%2C6000000
# electric https://www.bikewale.com/electric-bike/?sortField1=modelPopularity&sortOrder1=desc&budget=0%2C6000000&fuelType=807
driver.get("https://www.bikewale.com/electric-bike/?sortField1=modelPopularity&sortOrder1=desc&budget=0%2C6000000")
time.sleep(2)
# /html/body/div[1]/div[2]/div[2]/div[2]/section[2]/div/div/button
show_more_btn = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/div/button')

# Clicking to load all the product
while True:
    time.sleep(1)
    try:
        show_more_btn.click()
    except:
        break


contener = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/ul')

# Find all elements
elements = contener.find_elements(By.XPATH, r"./li")

def get_ftr(ele):
    dct=dict()
    text_ele=ele.find_elements(By.XPATH, './span')
    for txt in text_ele:
        text=txt.find_element(By.XPATH, './span[1]').text
        
        lst=text.split()
#         print(lst)
        if len(lst)>1:
            dct[lst[1]]=lst[0]
#         print(lst)
    return dct   

# print(len(elements))
# //*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/ul/li[1]/div/div[2]/div[2]/a/h3/span
# //*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/ul/li[1]/div/div[2]/div[2]/div[1]/div/div/p
# //*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/ul/li[1]/div/div[2]/div[2]/div[2]/span[1]/span[1]
# //*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/ul/li[1]/div/div[2]/div[2]/div[2]/span[2]/span[1]
for element in tqdm(elements):
    try:
        nm=element.find_element(By.XPATH, "./div/div[2]/div[2]/a/h3/span").text
        name.append(nm)
    except:
        nm=None
        name.append(nm)
    try:
        rt=element.find_element(By.XPATH, "./div/div[2]/div[2]/div[1]/div/div/p").text
        rating.append(rt)
    except:
        rt=None
        rating.append(rt)

    try:
        ele_dct=get_ftr(element.find_element(By.XPATH, "./div/div[2]/div[2]/div[2]"))
    except:
        print("There is some exception")
    # Defining Variables to None
    speed_kmph, single_charge_run_km, weight_kg, fll_charge_hr, cc_, milage_km, power_=None,None,None,None,None,None,None

    # If vales are present in geten dictionary then assign else none
    for var in ele_dct.keys():
        if var=='kmph' or var=='Kmph':
            speed_kmph=ele_dct[var]
        elif var=='km' or var=='Km':
            single_charge_run_km=ele_dct[var]
        elif var=='kg' or var=='Kg':
            weight_kg=ele_dct[var]
        elif var=='hrs' or var=='Hrs':
            fll_charge_hr=ele_dct[var]
        elif var=='cc':
            cc_=ele_dct[var]
        elif var=='kmpl' or var=='Kmpl':
            milage_km=ele_dct[var]
        elif var=='bhp' or var=='Bhp':
            power_=ele_dct[var]
            
    # Adding values to their list         
    speed.append(speed_kmph)
    charge.append(single_charge_run_km)
    wight.append(weight_kg)
    fl_charge.append(fll_charge_hr)
    cc.append(cc_)
    milage.append(milage_km)
    power.append(power_)
    try:
        price_=element.find_element(By.XPATH, './div/div[2]/div[2]/div[3]/span/span').text
        price.append(price_)
    except:
        price_=None
        price.append(price_)

    print(rt,nm,price_,speed_kmph, single_charge_run_km, weight_kg, fll_charge_hr, cc_, milage_km, power_)
# //*[@id="root"]/div[2]/div[2]/div[2]/section[2]/div/ul/li[1]/div/div[2]/div[2]/div[3]/span/span   

df=pd.DataFrame(list(zip(name, price, rating, speed, charge, wight, fl_charge, cc, milage, power)),
               columns=['name', 'price', 'rating', 'top_speed', 'riding_range', 
                        'weight', 'fl_charge_time', 'cc', 'milage', 'power'])

df.to_csv('C:\\Users\\Acer\\Desktop\\feynn labs\\bike.csv', index=False, encoding='utf-8-sig')
driver.quit()
# Bike wale weight present in some element and for some not present and class names not different
```
