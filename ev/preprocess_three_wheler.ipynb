```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv("/content/drive/My Drive/Colab Notebooks/Feynn Labs/ev/data/three_wheelers.csv")
df.head()
```

```python
df.info()
```

```python
def extract_price(text):
    if text=='Price Coming Soon':
        return 0.001
    elif text.find(',')!=-1:
      partten="(\d+.\d+|d+,\d+)"
#     print(text)
      match=re.search(partten,text)
      tmp=match.group(1).split(',')
      # print(tmp)
      return float(''.join(tmp))/100000

    partten="(\d+.\d+|d+)"
#     print(text)
    match=re.search(partten,text)
    return float(match.group(1))
df['price_new']=df['price'].apply(extract_price)
df['price_new']
```

```python
def extract_brand(text):
  try:
    return text.split()[0]
  except:
    return text
df['brand']=df['name'].apply(extract_brand)
df['brand'].unique()
```

```python
def extract_power(text):
    text=str(text)
    if text==None:
        return None
    wat = 'Watt' or 'W'
    partten="(\d+.\d+|\d+)"

    if re.search('Watt|W',text):
      # partten="(\d+)"
      match=re.search(partten,text)
      # print(round(float(match.group(1))/746,2))
      return round(float(match.group(1))/746,2)
      # print(text)
    else:
      # print(text)
      match=re.search(partten,str(text))
      try:
          return round(float(match.group(1)),2)
      except:
          return None
df['power_new']=df['power'].apply(extract_power)
# df['power_new'].unique()
```

```python
def extract_int(text):
    text=str(text)
    try:
        return float(text)
    except:
        partten="(\d+)"
#     print(text)
        match=re.search(partten,text)
        try:
            return float(match.group(1))
        except:
            return None
df['payload'].apply(extract_int).isna().sum()
```

```python
def extract_int(text):
    text=str(text)
    try:
        return float(text)
    except:
        partten="(\d+)"
#     print(text)
        match=re.search(partten,text)
        try:
            return float(match.group(1))
        except:
            return None
df['gvw_new']=df['gvw'].apply(extract_int)
df['milage_new']=df['milage'].apply(extract_int)
df['engine_new']=df['engine'].apply(extract_int)
df['fuel_tank_new']=df['fual_tank'].apply(extract_int)
df['payload_new']=df['payload'].apply(extract_int)
# df['payload_new']
```

```python
df_=df[df['price_new']!=0.001]
cols=['name', 'brand','price_new', 'power_new', 'gvw_new',
       'milage_new', 'engine_new', 'fuel_tank_new', 'payload_new', 'fuel_type']
df_new=df_[cols]
df_new.columns=['name', 'brand','price', 'power', 'gvw','milage', 'engine','fuel_tank', 'payload', 'fuel_type']
df_new.head()
```
