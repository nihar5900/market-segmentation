```python
import re
def extract_price(text):
    if text=='Price Coming Soon':
        return None
    partten="(\d+.\d+|d+)"
#     print(text)
    match=re.search(partten,text)
    return float(match.group(1))
df['price_new']=df['price'].apply(extract_price)
```

```python
def extract_power(text):
#     if text==None:
#         return None
    partten="(\d+)"
#     print(text)
    match=re.search(partten,text)
    try:
        return int(match.group(1))
    except:
        return None
df['power_new']=df['power'].apply(extract_power)
```

```python
def extract_fuel(text):
    try:
        return float(text)
    except:
        partten="Petrol.((\d+.\d+)|(\d+))L"
#     print(text)
        match=re.search(partten,text)
        try:
            return float(match.group(1))
        except:
            return None
def extract_int(text):
    try:
        return float(text)
    except:
        partten="(\d+)"
#     print(text)
        match=re.search(partten,text)
        try:
            return float(match.group(1))
        except:
            return None
df['gvw_new']=df['gvw'].apply(extract_int)
df['milage_new']=df['milage'].apply(extract_int)
df['engine_new']=df['engine'].apply(extract_int)
df['fuel_tank_new']=df['fual_tank'].apply(extract_fuel)
df['payload_new']=df['payload'].apply(extract_int)
# df['payload_new']
```
